# web scraping athena health

### Problema: Retornar o nome de uma série de páginas web e salvar em formato excel junto com seu respectivo id

Importando bibliotecas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Criando função que retorna o nome do título da página dado o id da clinica

```python
def get_clinic_name(clinic_id):
    url         = f'https://{clinic_id}.portal.athenahealth.com/'
    response    = requests.get(url)
    html        = response.text
    soup        = BeautifulSoup(html, 'html.parser')
    clinic_name = soup.find_all('h1')[-1].text.strip()
    return clinic_name 
```

loop para executar uma varredura em varios ids e retornar em uma lista

```python
start = 11000
end = 12710
```

```python
master_list = []

for clinic_id in range(start, end+1):
    data_dict = {}
    data_dict['clinic_id'] = clinic_id
    data_dict['clinic_name'] = get_clinic_name(clinic_id)
    if data_dict['clinic_name'] != 'Payment Confirmation' and data_dict['clinic_name'] != "Sorry, we can't find that practice. Make sure you typed the right address.":
        master_list.append(data_dict)
    print('{:.2f}%'.format((clinic_id-start)/end * 100))
```

Visualizando a lista

```python
master_list
```

transformando a lista em um dataframe

```python
df = pd.DataFrame(master_list)
```

salvando em formato csv

```python
df.to_csv('clinic_data.csv', index=False)
```
